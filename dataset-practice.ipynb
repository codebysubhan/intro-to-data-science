```python
# task 10: Retrieve rows where the employee belongs to "Sales" and has a Salary in the top 10% of all salaries.
floor_salary = df["Salary"].quantile(0.90)
print("employees from dept sales and salary in top 10%:")
df[(df["Department"] == "Sales") & (df["Salary"] >= floor_salary)]
```

```python
# task 11: Replace all "Inactive" values in the Status column with "Retired".
df["Status"] = df["Status"].replace("Inactive", "Retired")
```

```python
# task 12: Add a new column Bonus where values are 10% of the Salary column.
df["Bonus"] = df["Salary"]*0.10
```

```python
# task 13: .Update the Performance Score of employees in the "IT" department to 5 if it is NaN.

# df["Performance Score"]=df[(df["Department"] == "IT") & (df["Performance Score"].isna())].replace(np.nan, 5)
df.loc[(df["Department"] == "IT") & (df["Performance Score"].isna()), "Performance Score"] = 5
# df[(df["Department"] == "IT")]["Performance Score"].isna()
```

```python
# task 14: Modify the Experience of employees with less than 2 years to 2.
df["Experience"] = df["Experience"].apply(lambda x: 2 if x < 2 else x)
```

```python
# 15. Normalize the Salary column so that all values are between 0 and 1
df["Normalized Salary"] = (df["Salary"] - df["Salary"].min()) / (df["Salary"].max() - df["Salary"].min())
```

```python
# 16. Replace all Location values with "Remote" for employees with a Performance Score below 3
df.loc[df["Performance Score"] < 3, "Location"] = "Remote"
```

```python
# 17. Create a new column Experience Level based on the Experience
df["Experience Level"] = df["Experience"].apply(lambda x: "Junior" if x < 3 else ("Mid" if x < 6 else "Senior"))
```

```python
# 18. Add a new row with specific values using the append() method
new_row = {"Name": "Frank", "Department": "Sales", "Performance Score": 4, "Salary": 6000, "Location": "Remote", "Age": 28, "Experience": 2, "Joining Date": "2024-01-01", "Session": "Morning"}
df = df.append(new_row, ignore_index=True)
```

```python
# 19. Delete the column Session from the DataFrame
df = df.drop(columns=["Session"])
```

```python
# 20. Delete rows where the Age is greater than 60
df = df[df["Age"] <= 60]
```

```python
# 21. Add a new column Age Group based on Age
df["Age Group"] = df["Age"].apply(lambda x: "Young" if x < 35 else ("Middle-aged" if x < 50 else "Senior"))
```

```python
# 22. Identify all columns with missing values using the isna() method
missing_values = df.isna().sum()
```

```python
# 23. Fill the missing values in Performance Score with the average score of that column
df["Performance Score"].fillna(df["Performance Score"].mean(), inplace=True)
```

```python
# 24. Drop all rows with missing values using the dropna() method
df.dropna(inplace=True)
```

```python
# 25. Replace missing values in Performance Score with the median of the scores, grouped by Department
df["Performance Score"] = df.groupby("Department")["Performance Score"].transform(lambda x: x.fillna(x.median()))
```

```python
# 26. Identify rows where the employee is in the "IT" department and their Age is above the average Age of "IT" employees
avg_age_IT = df[df["Department"] == "IT"]["Age"].mean()
it_above_avg_age = df[(df["Department"] == "IT") & (df["Age"] > avg_age_IT)]
```

```python
# 27. Identify rows with duplicate Name and Joining Date combinations, and remove duplicates
df.drop_duplicates(subset=["Name", "Joining Date"], inplace=True)
```

```python
# 28. Remove rows where the Experience value is inconsistent with the Age (e.g., Experience > Age - 18)
df = df[df["Experience"] <= df["Age"] - 18]
```

```python
# 29. Sort the DataFrame by Salary in descending order
df = df.sort_values(by="Salary", ascending=False)
```

```python
# 30. Sort the DataFrame by Department and then by Joining Date in ascending order
df = df.sort_values(by=["Department", "Joining Date"])
```

```python
# 31. Rank all employees by Salary within their Department and add the rank as a new column
df["Salary Rank"] = df.groupby("Department")["Salary"].rank(ascending=False)
```

```python
# 32. Sort the dataset by Experience in descending order, but break ties using Performance Score in ascending order
df = df.sort_values(by=["Experience", "Performance Score"], ascending=[False, True])
```

```python
# 33. Create a new DataFrame containing only employees in "HR" with a Salary above 5000
hr_high_salary = df[(df["Department"] == "HR") & (df["Salary"] > 5000)]
```

```python
# 34. Filter out all rows where the Location is "Chicago"
df = df[df["Location"] != "Chicago"]
```

```python
# 35. Delete all rows where the Performance Score is below 2
df = df[df["Performance Score"] >= 2]
```

```python
# 36. Find the average Salary for each Department
avg_salary_by_dept = df.groupby("Department")["Salary"].mean()
```

```python
# 37. Calculate the total Salary of all employees grouped by Status
total_salary_by_status = df.groupby("Status")["Salary"].sum()
```

```python
# 38. Find the maximum Performance Score for employees in each Location
max_performance_by_location = df.groupby("Location")["Performance Score"].max()
```

```python
# 39. Calculate the total Salary paid to employees who joined in each year
df["Joining Year"] = pd.to_datetime(df["Joining Date"]).dt.year
total_salary_by_year = df.groupby("Joining Year")["Salary"].sum()
```

```python
# 40. Find the department with the highest average Performance Score
avg_performance_by_dept = df.groupby("Department")["Performance Score"].mean()
highest_avg_performance_dept = avg_performance_by_dept.idxmax()
```

```python
# 41. Count the number of employees in each Location grouped by their Session
employee_count_by_location_session = df.groupby(["Location", "Session"]).size()
```

```python
# 42. Group employees by Status and calculate the sum, mean, and standard deviation of their Salary
salary_stats_by_status = df.groupby("Status")["Salary"].agg(["sum", "mean", "std"])
```

```python
# 43. Use the apply() method to create a new column Monthly Salary by dividing the Salary by 12
df["Monthly Salary"] = df["Salary"].apply(lambda x: x / 12)
```

```python
# 44. Use the applymap() method to convert all string values in the DataFrame to uppercase
df = df.applymap(lambda x: x.upper() if isinstance(x, str) else x)
```

```python
# 45. Create a second dataset containing Department and Budget. Perform an inner join with the original dataset.
budget_data = {"Department": ["Sales", "IT", "HR"], "Budget": [100000, 200000, 150000]}
df_budget = pd.DataFrame(budget_data)
df_merged = pd.merge(df, df_budget, on="Department", how="inner")
```

```python
# 46. Merge the dataset with another dataset that includes additional information about employee ID and their recent Projects.
project_data = {"Employee ID": [1, 2, 3, 4], "Recent Project": ["Project A", "Project B", "Project C", "Project D"]}
df_projects = pd.DataFrame(project_data)
df_with_projects = pd.merge(df, df_projects, left_index=True, right_index=True, how="left")
```

```python
# 47. Perform a left join with a dataset containing Department and Team Lead details, adding Team Lead to the DataFrame.
team_lead_data = {"Department": ["Sales", "IT", "HR"], "Team Lead": ["Alice", "Bob", "Charlie"]}
df_team_lead = pd.DataFrame(team_lead_data)
df_with_team_lead = pd.merge(df, df_team_lead, on="Department", how="left")
```

```python
# 48. Create a DataFrame with employee ID and Project Count, then merge it with the original dataset.
project_count_data = {"Employee ID": [1, 2, 3, 4], "Project Count": [3, 2, 5, 4]}
df_project_count = pd.DataFrame(project_count_data)
df_with_project_count = pd.merge(df, df_project_count, left_index=True, right_index=True, how="left")
```

```python
# 49. Calculate the average Salary per project
df_with_project_count["Average Salary Per Project"] = df_with_project_count["Salary"] / df_with_project_count["Project Count"]
```

```python
# 50. Calculate the percentage increase in Salary for each employee compared to the average Salary in their Department
df["Avg Salary in Dept"] = df.groupby("Department")["Salary"].transform("mean")
df["Salary Percentage Increase"] = ((df["Salary"] - df["Avg Salary in Dept"]) / df["Avg Salary in Dept"]) * 100
```
